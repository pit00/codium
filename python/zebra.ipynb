```python
"""https://exercism.org/tracks/r/exercises/zebra-puzzle"""

import itertools


# 1. There are five houses.
orders = list(itertools.permutations(range(5)))

colors = ['yellow', 'blue', 'red', 'ivory', 'green']
people = ['Norwegian', 'Ukrainian', 'Englishman', 'Spaniard', 'Japanese']
drinks = ['water', 'tea', 'milk', 'juice', 'coffee']
pets = ['fox', 'horse', 'snails', 'dog', 'zebra']
positions = ['1', '2', '3', '4', '5']
hobbies = ['football', 'chess', 'dancing', 'painter', 'reading']

def solver(key, value):
    """Return a person"""

    # Colors
    for (yellow, blue, red, ivory, green) in orders:
        # 6. The green house is immediately to the right of the ivory house.
        if green != ivory + 1:
            continue # 6.

        # People
        for (Norwegian, Ukrainian, Englishman, Spaniard, Japanese) in orders:
            # 2. The Englishman lives in the red house.
            if Englishman != red:
                continue
            # 10. The Norwegian lives in the first house.
            if Norwegian != 0:
                continue
            # 15. The Norwegian lives next to the blue house.
            if abs(Norwegian - blue) != 1:
                continue

            # Drinks
            for (water, tea, milk, juice, coffee) in orders:
                # 4. The person in the green house drinks coffee.
                if coffee != green:
                    continue
                # 5. The Ukrainian drinks tea.
                if Ukrainian != tea:
                    continue
                # 9. The person in the middle house drinks milk.
                if milk != 2:
                    continue

                # Hobbies
                for (football, chess, dancing, painter, reading) in orders:
                    # 8. The person in the yellow house is a painter.
                    if painter != yellow:
                        continue
                    # 13. The person who plays football drinks orange juice.
                    if football != juice:
                        continue
                    # 14. The Japanese person plays chess.
                    if Japanese != chess:
                        continue

                    # Pets
                    for (fox, horse, snails, dog, zebra) in orders:
                        # 3. The Spaniard owns the dog.
                        if Spaniard != dog:
                            continue
                        # 7. The snail owner likes to go dancing.
                        if dancing != snails:
                            continue
                        # 11. The person who enjoys reading lives in the house next to the person with the fox.
                        if abs(reading - fox) != 1:
                            continue
                        # 12. The painter's house is next to the house with the horse.
                        if abs(painter - horse) != 1:
                            continue

                        houses = []
                        for i in range(5):
                            house = {
                                'color': colors[[yellow, blue, red, ivory, green].index(i)],
                                'person': people[[Norwegian, Ukrainian, Englishman, Spaniard, Japanese].index(i)],
                                'drink': drinks[[water, tea, milk, juice, coffee].index(i)],
                                'hobbie': hobbies[[football, chess, dancing, painter, reading].index(i)],
                                'pet': pets[[fox, horse, snails, dog, zebra].index(i)],
                            }
                            houses.append(house)
                        for house in houses:
                            if house[key] == value:
                                return house['person']


def drinks_water():
    """Which of the residents drinks water?"""

    return solver("drink", "water")

def owns_zebra():
    """Who owns the zebra?"""
    return solver("pet", "zebra")
```

```python
assert drinks_water() == "Norwegian"
```

```python
assert owns_zebra() == "Japanese"
```
